```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
%matplotlib inline 
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('Adaniport_new.csv')
```

```python
df = df.drop(['Date','Symbol','Deliverable Volume','Series','VWAP','Last','Turnover','Trades','%Deliverble',], axis='columns')
```

```python
inputs = df.drop(['Close'],axis='columns')
outputs = df['Close']
df
```

```python
df_stats = df.describe()
df_stats = df_stats.transpose()
df_stats
```

```python
def norm(x):
    return (x - df_stats['mean']) / df_stats['std']
df_normed = norm(df)
inputs_normed = df_normed.drop(['Close'], axis='columns')
outputs_normed = df_normed['Close']
```

```python
from sklearn.model_selection import train_test_split
inputs_train,inputs_test,outputs_train,outputs_test = train_test_split(inputs_normed,outputs,test_size=0.2)
```

```python
outputs_normed
```

```python
model = keras.Sequential([
    keras.layers.Dense(3,input_shape=[len(inputs.keys())],activation='relu'),
#     keras.layers.Dense(5,activation='relu'),
    
       
    keras.layers.Dense(1)
])

optimizer = tf.keras.optimizers.RMSprop(0.004)

model.compile(
    optimizer=optimizer,
    loss='mse',
    metrics=['mae','mse'] 
)
model.fit(inputs_normed,outputs,epochs=1100)
# model.fit(inputs,outputs,epochs=1000)
```

```python
model.evaluate(inputs_test,outputs_test)
```
